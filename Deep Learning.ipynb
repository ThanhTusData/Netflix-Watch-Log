 Dự đoán hành vi người dùng (User Behavior Prediction)
Tạo văn bản (Text Generation)
Chuyển đổi ngữ nghĩa (Semantic Mapping)

```python
# Import các thư viện cần thiết
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential, Model, load_model
from tensorflow.keras.layers import LSTM, Dense, Embedding, Dropout, Input
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
import joblib  # Để lưu và tải label encoder
import warnings
warnings.filterwarnings('ignore')

# Hàm để trực quan hóa quá trình huấn luyện
def plot_history(history, title):
    plt.figure(figsize=(12, 4))
    plt.plot(history.history['loss'], label='Loss')
    plt.plot(history.history['accuracy'], label='Accuracy')
    
    # Kiểm tra xem val_loss và val_accuracy có tồn tại không
    if 'val_loss' in history.history:
        plt.plot(history.history['val_loss'], label='Validation Loss')
    if 'val_accuracy' in history.history:
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    
    plt.title(title)
    plt.legend()
    plt.show()


# 1. Dự đoán hành vi người dùng (User Behavior Prediction)
df = pd.read_csv('All_SearchHistory.csv')

# Chuyển đổi nhãn 'Action' sang giá trị số
le = LabelEncoder()
df['Action'] = le.fit_transform(df['Action'])

# Lưu label encoder
joblib.dump(le, 'label_encoder.pkl')

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(df['Query Typed'], df['Action'], test_size=0.2, random_state=42)

# Mã hóa văn bản và padding chuỗi
tokenizer = Tokenizer(num_words=1000)
tokenizer.fit_on_texts(X_train)
max_len = max([len(seq) for seq in tokenizer.texts_to_sequences(X_train)])

X_train_seq = pad_sequences(tokenizer.texts_to_sequences(X_train), maxlen=max_len)
X_test_seq = pad_sequences(tokenizer.texts_to_sequences(X_test), maxlen=max_len)

# Xây dựng mô hình LSTM
model = Sequential([
    Embedding(input_dim=1000, output_dim=128, input_length=max_len),
    LSTM(64, return_sequences=False),
    Dropout(0.5),
    Dense(len(df['Action'].unique()), activation='softmax')
])

# Compile mô hình
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Huấn luyện mô hình
history = model.fit(X_train_seq, y_train, validation_data=(X_test_seq, y_test), epochs=10, batch_size=32)

# Trực quan hóa kết quả huấn luyện
plot_history(history, 'User Behavior Prediction')

# Lưu mô hình
model.save('user_behavior_model.h5')


# 3. Tạo văn bản (Text Generation)
corpus = df['Query Typed'].tolist()

# Mã hóa văn bản
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
total_words = len(tokenizer.word_index) + 1

input_sequences = []
for line in corpus:
    token_list = tokenizer.texts_to_sequences([line])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i + 1]
        input_sequences.append(n_gram_sequence)

# Padding sequences
max_sequence_length = max([len(x) for x in input_sequences])
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_length, padding='pre')
X, y = input_sequences[:, :-1], input_sequences[:, -1]

# One-hot encoding y
y = to_categorical(y, num_classes=total_words)

# Xây dựng mô hình LSTM cho tạo văn bản
model_text_gen = Sequential([
    Embedding(total_words, 100, input_length=max_sequence_length - 1),
    LSTM(150),
    Dense(total_words, activation='softmax')
])

model_text_gen.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Huấn luyện mô hình
history_text_gen = model_text_gen.fit(X, y, epochs=100, verbose=1)

# Trực quan hóa kết quả huấn luyện
plot_history(history_text_gen, 'Text Generation')

# Lưu mô hình
model_text_gen.save('text_generation_model.h5')


# 4. Chuyển đổi ngữ nghĩa (Semantic Mapping)
input_texts = df['Query Typed'].tolist()
target_texts = df['Displayed Name'].tolist()

# Mã hóa văn bản đầu vào và đầu ra
input_tokenizer = Tokenizer()
input_tokenizer.fit_on_texts(input_texts)
input_sequences = input_tokenizer.texts_to_sequences(input_texts)

target_tokenizer = Tokenizer()
target_tokenizer.fit_on_texts(target_texts)
target_sequences = target_tokenizer.texts_to_sequences(target_texts)

# Padding
max_encoder_seq_length = max([len(seq) for seq in input_sequences])
max_decoder_seq_length = max([len(seq) for seq in target_sequences])

encoder_input_data = pad_sequences(input_sequences, maxlen=max_encoder_seq_length)
decoder_input_data = pad_sequences(target_sequences, maxlen=max_decoder_seq_length)

# Xây dựng mô hình seq2seq
latent_dim = 256

encoder_inputs = Input(shape=(None,))
encoder_embedding = Embedding(input_dim=len(input_tokenizer.word_index) + 1, output_dim=latent_dim)(encoder_inputs)
encoder_outputs, state_h, state_c = LSTM(latent_dim, return_state=True)(encoder_embedding)
encoder_states = [state_h, state_c]

decoder_inputs = Input(shape=(None,))
decoder_embedding = Embedding(input_dim=len(target_tokenizer.word_index) + 1, output_dim=latent_dim)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True)
decoder_outputs = decoder_lstm(decoder_embedding, initial_state=encoder_states)
decoder_dense = Dense(len(target_tokenizer.word_index) + 1, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

model_seq2seq = Model([encoder_inputs, decoder_inputs], decoder_outputs)

model_seq2seq.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Padding target sequences
target_sequences_padded = pad_sequences(target_sequences, maxlen=max_decoder_seq_length, padding='post')
target_sequences_one_hot = to_categorical(target_sequences_padded, num_classes=len(target_tokenizer.word_index) + 1)

# Huấn luyện mô hình
history_semantic = model_seq2seq.fit([encoder_input_data, decoder_input_data], target_sequences_one_hot, epochs=100)

# Trực quan hóa kết quả huấn luyện
plot_history(history_semantic, 'Semantic Mapping')

# Lưu mô hình
model_seq2seq.save('semantic_mapping_model.h5')
```

Phát hiện bất thường (Anomaly Detection):

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import joblib

# 1. Đọc dữ liệu
df = pd.read_csv('All_ViewingActivity.csv')

# 2. Tiền xử lý dữ liệu
# Cập nhật hàm convert_duration
def convert_duration(duration):
    if 'days' in duration:
        days, time = duration.split('days')
        h, m, s = map(int, time.strip().split(':'))
        return int(days) * 86400 + h * 3600 + m * 60 + s  # 86400 giây trong một ngày
    else:
        h, m, s = map(int, duration.split(':'))
        return h * 3600 + m * 60 + s

df['Duration'] = df['Duration'].apply(convert_duration)

# Tạo cột 'is_anomaly' giả định rằng tất cả đều bình thường
df['is_anomaly'] = 0  # Giả định rằng không có bất thường

# Giả sử bạn muốn sử dụng Duration và một số cột khác cho mô hình
features = df[['Duration']]
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# 3. Xây dựng mô hình Autoencoder
model = Sequential([
    Dense(32, activation='relu', input_shape=(features_scaled.shape[1],)),
    Dense(16, activation='relu'),
    Dense(32, activation='relu'),
    Dense(features_scaled.shape[1], activation='sigmoid')
])

model.compile(optimizer='adam', loss='mse')

# 4. Huấn luyện mô hình (sử dụng dữ liệu không có anomalies để huấn luyện)
normal_data = features_scaled[df['is_anomaly'] == 0]
model.fit(normal_data, normal_data, epochs=50, batch_size=32, shuffle=True)

# 5. Lưu mô hình
model.save('autoencoder_model.h5')
joblib.dump(scaler, 'scaler.pkl')  # Lưu scaler để sử dụng lại
```
